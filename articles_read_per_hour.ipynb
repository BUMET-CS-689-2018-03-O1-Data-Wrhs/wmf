```python
%matplotlib inline
from query_remote_from_local_utils import *

#page views
query = """
CREATE TEMPORARY FUNCTION ua as 'org.wikimedia.analytics.refinery.hive.UAParserUDF';
SELECT y.views_per_hour, COUNT(*) as num_users FROM 
    (SELECT count(*) as views_per_hour
    FROM ellery.pageviews_no_spiders
      WHERE year = %(year)d
        AND month = %(month)d
        AND day = %(day)d
        AND hour = %(hour)d
        GROUP BY user_agent, ip, x_forwarded_for
    ) y
GROUP BY y.views_per_hour;   
"""

params = {'year':2014, 'month':12, 'day':1, 'hour':21}

query = query % params

d = query_hive_ssh(query, 'articles_per_hour.tsv')
```

```python
import pandas as pd
import numpy as np

def plot_cumulative(d, n = 200):
    n=200
    d.index = d.views_per_hour 
    d = d[:n]
    d['num_page_views'] = d['num_users'] * d['views_per_hour'] 
    
    d['percent']
    
    d['cumulative_num_page_views'] = np.cumsum(d['num_page_views'])
    d['cumulative_num_users'] = np.cumsum(d['num_users'])
    d['percent_cum_num_page_views'] = d['cumulative_num_page_views'] / d['num_page_views'].sum()
    d['percent_cum_num_users'] = d['cumulative_num_users'] / d['num_users'].sum()
    
    d[:10][['percent_num_page_views','percent_num_users']].plot()
    d[:10][['percent_num_page_views','percent_num_users']].plot()

```

```python
plot_cumulative(d, n = 200)
```

```python
unique_users = d['num_users'].sum()
pageviews = d['num_page_views'].sum()
print unique_users / float(pageviews - unique_users)
```
