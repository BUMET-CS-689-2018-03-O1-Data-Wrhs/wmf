```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
import pandas as pd
import matplotlib.pyplot as plt
pd.options.display.mpl_style = 'default'
import numpy as np
import os
from query_remote_from_local_utils import query_lutetium

    
def get_utm_keys(start, end):
    query = """
    SELECT
    DATE_FORMAT(ts, '%%Y-%%m-%%d') AS timestamp,
    utm_key AS impressions_seen,  
    COUNT(*) as count 
    FROM drupal.contribution_tracking  
    WHERe ts > %(start)s 
    AND ts < %(end)s 
    AND contribution_id is not NULL 
    AND utm_medium = 'sitenotice' 
    GROUP BY timestamp, utm_key 
    ORDER BY timestamp, utm_key;
    """ 
            
    d = query_lutetium(query, {'start':start, 'end':end})
    d.index  = d['timestamp'].map(lambda t: pd.to_datetime(str(t)))
    del d['timestamp']
    return d

start = '20131201000000'
end =   '20131231230000'

d = get_utm_keys(start, end)
```

```python

d1 = d['2013-12-03': '2013-12-07'].groupby('impressions_seen').sum()
d2 = d['2013-12-07': '2013-12-015'].groupby('impressions_seen').sum()
d3 = d['2013-12-17': '2013-12-29'].groupby('impressions_seen').sum()
d4 = d['2013-12-31': ].groupby('impressions_seen').sum()
d1_2 = d['2013-12-03': '2013-12-15'].groupby('impressions_seen').sum()

dall = d.groupby('impressions_seen').sum()
```

```python
def nice_hist(d):
    d = d.sort()
    fig = plt.figure()
    ax = fig.add_subplot(111)
    d = d.ix[d.index < 50]
    fig.set_size_inches(15, 8)
    width = 0      # the width of the bars

    ind = np.arange(d.shape[0])  # the x locations for the groups
    a_rects = ax.bar(ind, d['count'], align='center', facecolor = 'yellow', edgecolor='gray')
    ax.set_xticks(ind+width/2)
    ax.set_xticklabels( [int(i) for i in d.index])
    ax.set_xlabel('impressions seen before donating')
    ax.set_ylabel('count')
    


    
```
