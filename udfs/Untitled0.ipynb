```python
import re
production_sites = [
'wikipedia',
'wiktionary',
'wikibooks',
'wikinews',
'wikiquote',
'wikisource',
'wikiversity',
'wikivoyage',
'wikimedia',
'wikidata']
```

```python
regex = '(www\\.)?'\
+ '(((?:(?!m\\.|zero\\.|wap\\.|mobile\\.)[^.])*)\\.)?' \
+ '((m|zero|wap|mobile)\\.)?'\
+ '('+'|'.join(production_sites)+')\\.org(:80)?'
print regex
b = re.compile(regex)
```

```python
for line in open('hosts.txt'):
    host = line.split('\t')[0].strip()
    print line
    m = re.match (b, line)
    if m:
        code = m.group(3)
        #if code:
        #    code = code[:-1]
            
        access_method = m.group(5)
        if access_method:
            access_method = access_method
        else:
            access_method = 'd'
        
        site_code = m.group(6) 
        print code,access_method, site_code 
    else:
        print 'no match'
    print '\n\n'
```
