```python
from db_utils import query_lutetium, get_time_limits
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from ggplot import *

```

```python
def get_contacts():


    query = """
    SELECT
    civicrm_contribution.total_amount,
    civicrm_country.iso_code,
    civicrm_contribution.receive_date,
    civicrm_email.email,
    civicrm_address.street_address,
    civicrm_address.postal_code,
    civicrm_address.city,
    civicrm_contact.first_name,
    civicrm_contact.last_name,
    contribution_tracking.utm_medium
    FROM
    civicrm.civicrm_contribution,
    civicrm.civicrm_address,
    civicrm.civicrm_country,
    civicrm.civicrm_email,
    civicrm.civicrm_contact,
    drupal.contribution_tracking
    
    WHERE 
    civicrm_contribution.id = drupal.contribution_tracking.contribution_id
    AND civicrm_contribution.contact_id = civicrm_address.contact_id
    AND civicrm_address.country_id = civicrm_country.id
    AND civicrm_contact.id = civicrm_contribution.contact_id
    AND civicrm_contribution.contact_id = civicrm_email.contact_id
    AND civicrm_address.is_primary = 1
    AND civicrm_country.iso_code = 'US'
    ORDER BY civicrm_contact.first_name;
    """
    #fraction_of_first_time_donors
    return query_lutetium(query, {})
    
```

```python
d_orig = get_contacts()
```

```python
d_orig.to_csv('donors.csv')
```

```python
d_orig['id'] = d_orig.index
print d_orig.shape
```

```python
d = d_orig
```

```python
# We first merge all contacts with the same email address
import numpy as np
def merge(d):
    d.id = d.id.min()
    return d
    
d = d.groupby('email').apply(merge)
```

```python
print d.shape
print len(set(d.id))
d = d.fillna('')
```

```python
# Lets map all strings to lower case and standardize abbreviations
```

```python
def lower(s):
    try:
        return s.lower()
    except:
        return s

d['city'] = d['city'].apply(lower)
d['first_name'] = d['first_name'].apply(lower)
d['last_name'] = d['last_name'].apply(lower)
d['street_address'] = d['street_address'].apply(lower)


    
#d = d.groupby(['first_name', 'last_name','street_address', 'city' ]).apply(merge)
```

```python
print d.shape
print len(set(d.id.apply(int)))
```

```python
def map_street_abbreviations(s):
    try:
        s = s.replace('  ', ' ')
        s = s.replace('.', '')
        
        s = s.replace(' ct', ' court')
        s = s.replace(' street', ' st')
        s = s.replace(' str', ' st')
        s = s.replace(' boulevard', ' blvd')
        s = s.replace(' lane', ' ln')
        s = s.replace(' av', ' ave')
        s = s.replace(' avenue', ' ave')
        s = s.replace(' drive', ' dr')
        s = s.replace(' road', ' rd')
        return s
        
    except:
        return s
    
d['street_address'] = d['street_address'].apply(map_street_abbreviations)
d = d.groupby(['first_name', 'last_name','street_address', 'city' ]).apply(merge)
print d.shape
print len(set(d.id))
```

```python
#Number of Unique Donors per year
```

```python
d['year'] = d['receive_date'].apply(lambda x: x.year)
max_year = 2015
min_year = 2008
years = range(min_year, max_year)
donor_sets = { y:set(d[d.year == y].id) for y in years}
donors_per_year = [len(donor_sets[i]) for i in years]

df = pd.DataFrame()
df['year'] = years
df['year'] = df['year'] -2000
df['donors_per_year'] = donors_per_year
df
```

```python
ggplot(df,aes(x='year', y='donors_per_year')) + geom_line() \
    + ggtitle("Number of Unique Donors per Year")
```

For the years 2008-2014, this plot shows the number of unique US donors.

```python
#The fraction of donors that gave in a  specific year that gave this year
```

```python
donors_giving_in_2014 = []
for y in range(min_year, max_year-1):
    repeats = donor_sets[y].intersection(donor_sets[max_year-1])
    donors_giving_in_2014.append(len(repeats)/float(len(donor_sets[y])))
    
donors_giving_in_2014.append(1)
    
df['fraction_of_donors_giving_in_2014'] = donors_giving_in_2014
```

```python
ggplot(df[:-1],aes(x='year', y='fraction_of_donors_giving_in_2014')) + geom_line() \
+ ggtitle("Fraction of donors who gave again in 2014")
```

For the years 2008-2014, this plot shows what fraction of people who donated in a particular year donated again in 2014. For example, 35% of people who donated in 2008 donated again in 2014.

```python
#Fraction of first time donors
```

```python
fraction_of_first_time_donors = [1, ]
for y in years[1:]:
    this_year = donor_sets[y]
    last_years = set(d[d.year < y].id)
    difference = this_year.difference(last_years)
    fraction_of_first_time_donors.append(len(difference)/float(len(this_year)))
```

```python
df['fraction_of_first_time_donors'] = fraction_of_first_time_donors
```

```python
ggplot(df,aes(x='year', y='fraction_of_first_time_donors')) + geom_line() \
+ ggtitle("Fraction of first time donors")
```

For years 2008-2014, this plot shows what fraction of donors are giving for the first time. For example, in 2013, we 65% of donors are new, meaning they did not give from 2008-2012.
