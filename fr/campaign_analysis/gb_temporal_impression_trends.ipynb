#Weekly Impressions: GB Desktop

```python
config = {
            'start' : '2014-12-01 00:00',
            'stop' : '2014-12-08 000:00',
            'ylabel': 'impressions per hour',
            'hours': 1
}

plot_by_time(imp, {'All': '.* C14_en5C_dec_dsk_FR'}, **config )
plot_by_time(imp, {'Large Desk': '.*5C_lg.* C14_en5C_dec_dsk_FR'}, **config )
#plot_by_time(imp, {'Top Desk': '.*5C_tp.* C14_en5C_dec_dsk_FR'}, **config )
```

"All" refers to any banner impression (proxy for pageviews when the campaign is at 100%). "Large Desk" refers to impressions of the Full screen banner, which was only show once per user. You can think of this being the number of people visiting for first time since Dec 2. There is not quite, a full week, but it looks like impressions fall over the course of the week, reaching a low point on Saturday. Lets look at another week:

```python
config = {
            'start' : '2014-12-08 00:00',
            'stop' : '2014-12-15 000:00',
            'ylabel': 'impressions per hour',
            'hours': 1
}

plot_by_time(imp, {'All': '.* C14_en5C_dec_dsk_FR'}, **config )
plot_by_time(imp, {'Large Desk': '.*5C_lg.* C14_en5C_dec_dsk_FR'}, **config )
#plot_by_time(imp, {'Top Desk': '.*5C_tp.* C14_en5C_dec_dsk_FR'}, **config
```
