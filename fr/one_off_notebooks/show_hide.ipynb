```python
import os
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
query ="select count(*) as count, \
parse_url(concat('http://bla.org/woo/', uri_query), 'QUERY', 'result') as result,  \
parse_url(concat('http://bla.org/woo/', uri_query), 'QUERY', 'banner') as banner, \
parse_url(concat('http://bla.org/woo/', uri_query), 'QUERY', 'reason') as reason, \
parse_url(concat('http://bla.org/woo/', uri_query), 'QUERY', 'campaign') as campaign, \
substr(dt, 0, 14) as minute \
from wmf_raw.webrequest \
where  year = 2014 and month = 11 and day = 5 and hour = 14 \
and uri_path = '/wiki/Special:RecordImpression' \
and parse_url(concat('http://bla.org/woo/', uri_query), 'QUERY', 'country') = 'US' \
and parse_url(concat('http://bla.org/woo/', uri_query), 'QUERY', 'project') = 'wikipedia' \
and parse_url(concat('http://bla.org/woo/', uri_query), 'QUERY', 'device') = 'desktop' \
and parse_url(concat('http://bla.org/woo/', uri_query), 'QUERY', 'uselang') = 'en' \
group by parse_url(concat('http://bla.org/woo/', uri_query), 'QUERY', 'result'), \
parse_url(concat('http://bla.org/woo/', uri_query), 'QUERY', 'banner'), \
parse_url(concat('http://bla.org/woo/', uri_query), 'QUERY', 'reason'), \
parse_url(concat('http://bla.org/woo/', uri_query), 'QUERY', 'campaign'), \
substr(dt, 0, 14);"

    
def query_hive(query):
    cmd = """ssh stat1002.eqiad.wmnet "hive -e \\" """ +query+ """ \\" "> query_hive.csv"""
    os.system(cmd)
    d = pd.read_csv('query_hive.csv',  sep='\t')
    #os.system('rm query_hive.csv',)
    return d
```

```python
d_raw = query_hive(query)
```

```python
d = d_raw.fillna('none')
d['minute'] = d['minute']+"0"
#d['minute']  = d['minute'].map(lambda t: pd.to_datetime(str(t)))
```

```python
def plot_show_hide(d, banner):
    d = d.ix[d['banner'] == banner]
    d.index = pd.MultiIndex.from_tuples(zip(d['minute'], d['result'], d['reason']))
    d = d.sort()
    d = d[['count',]]
    d = d.unstack(level = [1, 2])
    d.columns = d.columns.droplevel(0)
    
    fig = plt.figure(figsize=(10, 6), dpi=80)
    ax = fig.add_subplot(111)
    d.plot(ax = ax, kind='bar', stacked=True, legend = True)
    ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

```python

plot_show_hide(d, 'B14_1030_enUS_dsk_tx_n2')
```

```python
plot_show_hide(d, 'B14_1015_enUS_dsk_tx_y2')
```
