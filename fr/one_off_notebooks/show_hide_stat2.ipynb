```python
import os
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import datetime
```

```python
def query_stat2(query, file_name):
        cmd = """ssh stat1002.eqiad.wmnet "mysql --defaults-file=/etc/mysql/conf.d/analytics-research-client.cnf -h analytics-store.eqiad.wmnet -u research -e \\" """ +query+ """ \\" --socket  /tmp/mysql.sock  "> """+ file_name
        os.system(cmd)
        d = pd.read_csv(file_name,  sep='\t')
        os.system('rm ' + file_name)
        return d
```

```python
def get_record_impression_data(t):
    
    
    query = "select \
    sum(count) as cnt, dt, result, reason \
    from staging.record_impression \
    where %s = '%s' \
    group by result, reason, dt \
    order by dt;"
    
    query = query % t
    d = query_stat2(query, "_impressions.csv")
    d.index = pd.to_datetime(d['dt'])
    d = d.fillna('none')

    return d



    
```

```python
def plot_show_hide(t, stop = '2050', minutes = 60):
    d = get_record_impression_data(t)
    
    d = d[:stop]
    #f = 
    d['dt'] = pd.to_datetime(d['dt'])
    d['dt'] = d['dt'].apply(lambda tm : tm - datetime.timedelta( minutes=(60 * tm.hour + tm.minute) % minutes))
    d.index = pd.MultiIndex.from_tuples(zip(d['dt'], d['result'], d['reason']))
    d = d[['cnt',]]
    d = d.groupby(d.index).sum()
    d.index = pd.MultiIndex.from_tuples(d.index)

    d = d.unstack(level = [1, 2])
    d.columns = d.columns.droplevel(0)
    
    fig = plt.figure(figsize=(24, 12), dpi=80)
    ax = fig.add_subplot(111)
    d.plot(ax = ax, kind='bar', stacked=True, legend = True)
    ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    
    return d
```

```python
d = plot_show_hide(('banner', 'B14_1105_enUS_dsk_twn1'),stop = '2014-11-5 17:15', minutes = 1 )
```
