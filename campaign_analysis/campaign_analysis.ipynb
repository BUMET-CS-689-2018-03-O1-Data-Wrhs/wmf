#Donation Rate: top vs full

```python
config = {
            'hours' :2, 
            'start': '2014-12-2 20',
        }

plot_rate_by_time(don, imp, desk_regs, **config)
```

```python
config = {
            'hours' :2, 
            'start': '2014-12-8 00',
        }

plot_rate_by_time(don, imp, mob_regs, **config )
```

The donation rates are vastly different! The top screen banner has had the same design for the last 4 days. Its donation rate looks pretty steady at a rate of 1:1000. It is strange that we do not see a daily pattern. At this rate, we will need roughly 9 million impressions per banner to detect 5% gains. The donation rate of full screen banners is high but diminishing. This is may be because we are getting the set of readers that don't visit wikipedia very often and are less likely to donate. There is also a daily pattern.
