This weeks again shows an increase from Monday to Thurday, a dip on Friday and then an increase on the wekeend. For this weekend the increase was pretty small compared to other weeks, maybe due to holidays.

There are also strong hourly trends. Remember the data is for the entire US. An east coasters lunch-time is a west-coaster's morning. Ideally, we would adjust based on city, which requires ip and geolocation. We could also try to find a small country that is similar to the US as a proxy. 

The donation rate increases rapidly from 1-5 PST and falls more gradually from 5-2 PST.  Basically, the rate drop over the course of the day.
