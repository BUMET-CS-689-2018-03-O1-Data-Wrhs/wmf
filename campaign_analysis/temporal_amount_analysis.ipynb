```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
import pandas as pd
import matplotlib.pyplot as plt
pd.options.display.mpl_style = 'default'

from amount_analysis import *
```

```python
start = '2014-11-28 00:00'
stop = '2015-01-03 00:00'

don = get_donations(start, stop, 'C14_en5C_dec_dsk_FR')
```

```python
desk_regs = {
    'Large Desk': '.*5C_lg.* C14_en5C_dec_dsk_FR',
    'Top Desk': '.*5C_tp.* C14_en5C_dec_dsk_FR',
}
```

```python
don['weekday'] = don.index.weekday
don['hour'] = don.index.hour
```

```python
don_us = don[don['country']=='US']
don_gb = don[don['country']=='GB']
don_ca = don[don['country']=='CA']
don_au = don[don['country']=='AU']
```

## Average Donation by Day of Month for US, GB, CA, AU

The average donation increases over the course of the campaign and the top screen banner has a slightly higher average donation.

```python
plot_central_tendency(don_us,desk_regs, stop = '2014-12-31',title = 'US', amount_limit = 100.0, hours = 24)
plot_central_tendency(don_gb,desk_regs, stop = '2014-12-31',title = 'GB', amount_limit = 100.0, hours = 24)
plot_central_tendency(don_ca,desk_regs, stop = '2014-12-31', title = 'CA', amount_limit = 100.0, hours = 24)
plot_central_tendency(don_au,desk_regs, stop = '2014-12-31',title = 'AU', amount_limit = 100.0, hours = 24)
```

```python
plot_central_tendency(don_us,desk_regs, start = '2014-12-3', stop = '2014-12-16', amount_limit = 100.0, index = 'weekday', title = 'US')
plot_central_tendency(don_gb,desk_regs, start = '2014-12-3', stop = '2014-12-16', amount_limit = 100.0, index = 'weekday', title = 'GB')
plot_central_tendency(don_ca,desk_regs, start = '2014-12-3', stop = '2014-12-16', amount_limit = 100.0, index = 'weekday', title = 'CA')
plot_central_tendency(don_au,desk_regs, start = '2014-12-3', stop = '2014-12-16', amount_limit = 100.0, index = 'weekday', title = 'AU')
```
